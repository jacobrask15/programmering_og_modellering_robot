```python
import matplotlib.pyplot as plt
import numpy as np


with open("test.txt", "r") as f:

    testOut2 = f.read()
    stringData = testOut2
    stringData = stringData.replace("mk_", "")


    data5 = eval(stringData)


    # Extract time and values, ignoring ON/OFF and direction for this plot
    times = [point[0] for point in data5]
    angles = [point[1] for point in data5]
    velocities = [point[2] for point in data5]
    accelerations = [point[3] for point in data5]

    # Create the plot
    #plt.ylim(-1,1)
    #plt.xlim(0, 3)
    plt.plot(times, accelerations, '-o', label='Acceleration', color='red', markersize=3)
    plt.plot(times, velocities, '-o', label='Velocity', color='yellow', markersize=3)
    plt.plot(times, angles, '-o', label='Angle', color='green', markersize=3)
    plt.xlabel('Time')
    plt.ylabel('Angle, Velocity, Acceleration')
    plt.title('Target: 0,6 rad')
    plt.legend()
    plt.grid(True)
    plt.show()

```
