```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make a window with four plots plotting time vs angle for 4 scenarios
plot1time = np.array([14, 4, 4, 0, 0, 0])
plot1time = plot1time[::-1]
plot1angle = np.array([45, 45, 45, 5, 5, 5])
plot1angle = plot1angle[::-1]
plot2time = np.array([30, 16, 16, 7, 7, 2, 2, 0, 0])
plot2time = plot2time[::-1]
plot2angle = np.array([-75, 65, 65, -25, -25, 25, 25, 5, 5])
plot2angle = plot2angle[::-1]
plot3time = np.array([18, 8, 8, 7, 7, 5, 5, 0, 0, 0])    
plot3time = plot3time[::-1]
plot3angle = np.array([45, 45, 45, 35, 35, 55, 55, 5, 5, 5])
plot3angle = plot3angle[::-1]
plot4time = np.array([56, 56, 56, 56, 56, 54, 54, 54, 50, 50, 50, 44, 44, 44, 36, 36, 36, 26, 26, 26, 14, 14, 14, 0, 0])
plot4time = plot4time[::-1]
plot4angle = np.array([205, 205, 205, 205, 205, 225, 225, 225, 185, 185, 185, 245, 245, 245, 165, 165, 165, 265, 265, 265, 145, 145, 145, 5, 5])
plot4angle = plot4angle[::-1]

# make a window with four plots plotting time vs angle for 4 scenarios
fig, ax = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.7, hspace=0.7)
#insert target line
ax[0, 0].axhline(y=45, color='g', linestyle='--')
ax[0, 1].axhline(y=45, color='g', linestyle='--')
ax[1, 0].axhline(y=45, color='g', linestyle='--')
ax[1, 1].axhline(y=45, color='g', linestyle='--')
ax[0, 0].plot(plot1time, plot1angle, 'o--')
ax[0, 0].set_title('Scenario 1')
ax[0, 0].set_xlabel('Time (s)')
ax[0, 0].set_ylabel('Angle (deg)')
ax[0, 1].plot(plot2time, plot2angle, 'o--')
ax[0, 1].set_title('Scenario 2')
ax[0, 1].set_xlabel('Time (s)')
ax[0, 1].set_ylabel('Angle (deg)')
ax[1, 0].plot(plot3time, plot3angle, 'o--')
ax[1, 0].set_title('Scenario 3')
ax[1, 0].set_xlabel('Time (s)')
ax[1, 0].set_ylabel('Angle (deg)')
ax[1, 1].plot(plot4time, plot4angle,   'o--')
ax[1, 1].set_title('Scenario 4')
ax[1, 1].set_xlabel('Time (s)')
ax[1, 1].set_ylabel('Angle (deg)')
plt.show()
```
